## Matplotlib Example Workflow

```python
# 0 importmathplotlib and get ready for plotting jupyer

%matplotlib inline 
import matplotlib.pyplot as plot 

# some data
x = [2,4,6,8]
y = [11,22,33,44]

# setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3 plot the data
ax.plot(x,y)

#4 customize plot 
ax.set(title="Simple Plot", 
      xlabel="x-axis",
      ylabel="y-axis")

# 5 save and show (you save the whole figure)

fig.savefig("../images/sample-plot.png")
```

### Which one is the best to use? (plplot vs matplotlib OO method)

* notes to self
* plotting something quickly, pyplot method is good
* plotting more advance, use the OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
len(over_50)
```

```python
over_50.plot(kind="scatter", 
            x="age",
            y="chol",
            c="target");
```

## OO method 

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);

#ax is where the data is

ax.set_xlim([45,80])

# set the limit on the x ax, lower 45, upper 80
```

```python
## OO method again is better

fig, ax = plt.subplots(figsize=(10,6))

# plot the data 
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# customize the plot

ax.set(title="Heart Disease and Cholesterol Levels ", 
      xlabel="Age",
      ylabel="Cholesterol");

# ad legends and find from C permaters and colours 
ax.legend(*scatter.legend_elements(), title="Target");

# how to add horizontal line 

ax.axhline(over_50["chol"].mean(),
          linestyle="--");

# add data to the ax1

scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

# custom ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate");
```

```python
## Another one with more figures using OO

# style 

plot.style.use("seaborn-whitegrid")

# setup

fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1, 
                              figsize=(10,10))

#cmap style on documentation [winter, summer autume, spring... and lots more]
scatter = ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                     cmap="winter") 

#cutomize ax0
ax0.set(title="Heart Disease and Cholesterol Levels ", 
      xlabel="Age",
      ylabel="Cholesterol");

# ad legends and find from C permaters and colours 
ax0.legend(*scatter.legend_elements(), title="Target");

# how to add horizontal line 

ax0.axhline(over_50["chol"].mean(),
          linestyle="--");
# change x ax limit 
ax0.set_xlim([50,80]);


# add data to the ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="summer")
# custom ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate");

ax1.legend(*scatter.legend_elements(), title="Target")

ax1.axhline(over_50["thalach"].mean(),
           linestyle="--");

# change ax1 axis limit
ax1.set_xlim([50,80]);

# change ax1 y axis limit
ax1.set_ylim([60,200]);

# add title to figure

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
